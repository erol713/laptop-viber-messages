```python
page = open(r'C:\Users\HP\Desktop\message history\html\Viber\Amir Ligata.html', encoding="utf-8")
```

```python
soup = BeautifulSoup(page.read())
content = soup.find_all('p', class_=["date", "triangle-isosceles", "triangle-isosceles2"])

message = []
name = soup.findAll('p')[0].text.strip()

for i in content:    
    if ('class="date"' in str(i)):       
        message.append(i.text.strip()[:10])        
        message.append(i.text.strip()[11:])    
        
    if ('class="triangle-isosceles"' in str(i)):
        message.append(name)
        message.append(i.text.strip())   
        
    if ('class="triangle-isosceles2"' in str(i)):
        message.append("Me")
        message.append(i.text.strip())       
```

```python
message = message[2:]

data = [message[n:n+4] for n in range(0, len(message), 4)]
```

```python
df = pd.DataFrame(data, columns=["Date", "Time of day", "Sender", "Content"])
#df.to_csv(name + ".csv", encoding='utf-8', index=False)
```
